```python
import sqlite3
```

```python
%load_ext sql
%sql sqlite:///q1.db
```

```python
%%sql
SELECT M1.rfid, M1.transaction_date AS 'Purchase Date', M2.transaction_date AS 'Return Date',M1.type_detail,M2.type_detail,
Cast ((M1.transaction_date - M2.transaction_date)AS Integer) AS 'Days between Purchase and Return'
FROM q1 AS M1 JOIN q1 AS M2 ON M1.rfid =M2.rfid
WHERE M1.type_detail='sale' AND M2.type_detail='return' AND M1.transaction_date<M2.transaction_date
```
